### Introducción

Puede haber situaciones en las que no tengamos ninguna variable respuesta que predecir. Estos problemas, sin ninguna variable objetivo explícita, se conocen como problemas de aprendizaje no supervisado. En estos problemas sólo tenemos las variables independientes.


image.png

En estos casos, intentamos clasificar los datos en un conjunto de grupos. Estos grupos se conocen como clusters y el proceso de hacer clusters se conoce como clustering.

image.png

Esta técnica se utiliza generalmente para agrupar una población en diferentes grupos. Algunos ejemplos comunes incluyen segmentar clientes, agrupar documentos similares, recomendar canciones o películas similares, etc. Existen MUCHAS más aplicaciones del aprendizaje no supervisado sumamente relacionadas a nuestra vida cotidiana.

Ahora bien, hay varios algoritmos que nos ayudan a hacer estos clusters. Los algoritmos de clustering más utilizados son K-means y Hierarchical clustering.

### Analisis de conglomerados

#### Objetivo

* Identicar grupos relativamente homogeneos con base en un conjunto de variables especificas.
* Clasificar a las observaciones en grupos donde cada grupo o conglomerado sea homogeneo internamente y heterogeneos entre ellos.

![](https://i.stack.imgur.com/cIDB3.png)

El criterio básico para cualquier agrupación en clúster es la **distancia**. Los objetos que están cerca unos de otros deben pertenecer al mismo clúster, y los objetos que están lejos unos de otros deben pertenecer a clústeres diferentes. Para un conjunto determinado de datos, los clústeres que se construyen dependen de la especificación de los parámetros siguientes:

* Método de clúster define las reglas para la formación de clúster. Por ejemplo, al calcular la distancia entre dos clústeres, puede utilizar el par de objetos más próximos entre clústeres o el par de objetos más lejanos, o un compromiso entre estos métodos.
* Medida define la fórmula para calcular la distancia. Por ejemplo, la medida de distancia euclídea calcula la distancia como una "línea recta" entre dos clústeres. Las medidas de intervalo asumen que las variables son de escala; las medidas de recuento asumen que son numéricas discretas; y las medidas binarias asumen que sólo toman dos valores.
* La estandarización le permite igualar el efecto de las variables medidas en diferentes escalas. De otra forma las variables con escalas de mayor magnitud opacarian todas las otras.

#### Algoritmos de clustering

##### **Clustering jerárquico:**

En la minería de datos y la estadística, el clustering jerárquico (también llamado análisis jerárquico de clusters o HCA) es un método de análisis de clusters que busca construir una **jerarquía de clusters**.

Es muy interesante llevar a cabo este analisis cuando se desea observar el orden en que se generan los clusteres, este metodo permite obtener el **historial de conglomeracion** o **dendrograma**.

![](https://www.datanovia.com/en/wp-content/uploads/dn-tutorials/003-hierarchical-clustering-in-r/figures/002-agglomerative-clustering-cutree-cut-dendrogram-1.png)

Las estrategias de clustering jerárquico se dividen generalmente en dos tipos:

1. **Aglomerativo:** Se trata de un enfoque "ascendente": cada observación comienza en su propio conglomerado, y los pares de conglomerados se fusionan a medida que se asciende en la jerarquía. Parte de la idea que tenemos tantos clusteres como datos y va agrupando cada uno de los clusteres hasta finalizar en uno unico.

2. **Divisivo:** Se trata de un enfoque "descendente": todas las observaciones comienzan en un conglomerado y las divisiones se realizan de forma recursiva a medida que se desciende en la jerarquía. A partir de un cluster unico se analiza como se va dividiendo para dar origen a un cluster por dato.





### Fundamentos de la agrupación jerárquica

- Creación de una matriz de distancias mediante enlace
    - ``method``: cómo calcular la proximidad de los clusters
    - ``metric``: métrica de distancia
    - ``ordenación_óptima``: ordenar los puntos de datos

### Diferentes formas de medir la distancia entre dos conglomerados

#### **Simple**: basado en los dos objetos más cercanos

![](https://i0.wp.com/dataaspirant.com/wp-content/uploads/2020/12/10-Simple-Linkage-Method.png?resize=768%2C594&ssl=1)

* El método de Enlace Simple también se conoce como método de Enlace Mínimo (MIN).

En el método de Enlace Simple, la distancia entre dos clusters se define como la distancia mínima entre un objeto (punto) de un cluster y un objeto (punto) del otro cluster. Este método también se conoce como método del vecino más próximo.

* Ventajas e inconvenientes del método Simple Linkage

**Ventajas del método Simple Linkage**

* Los métodos Simple Linkage pueden manejar formas no elípticas.
* Los algoritmos de Simple Linkage son los mejores para capturar clusters de diferentes tamaños.

**Contras del método Simple Linkage**

* Los métodos Simple Linkage son sensibles al ruido y a los valores atípicos.
* Esto significa que los métodos Simple Linkage no pueden agrupar clusters correctamente si hay ruido entre los clusters.

#### **Completo**: basado en los dos objetos más alejados

![](https://i0.wp.com/dataaspirant.com/wp-content/uploads/2020/12/11-Complete-Linkage-Method.png?resize=768%2C612&ssl=1)

* El método de Enlace completo también se conoce como método de Enlace máximo (MAX).

En la técnica de Enlace completo, la distancia entre dos clusters se define como la distancia máxima entre un objeto (punto) de un cluster y un objeto (punto) del otro cluster.

Este método también se conoce como método del vecino más lejano.

* Pros y contras del método Complete Linkage

**Ventajas del método Complete Linkage**

* Los algoritmos Complete Linkage son menos susceptibles al ruido y a los valores atípicos.
* Esto significa que el método Complete Linkage también funciona bien a la hora de separar conglomerados si hay algún ruido entre los conglomerados.

**Desventajas de la vinculación completa**

* Los métodos de enlace completo tienden a romper los conglomerados grandes.
* Complete Linkage se inclina hacia los clusters globulares.

#### **Promedio**: basado en la media aritmética de todos los objetos

![](https://i0.wp.com/dataaspirant.com/wp-content/uploads/2020/12/12-Average-Linkage-Method.png?resize=768%2C661&ssl=1)

* En la técnica Average Linkage, la distancia entre dos clusters es la distancia media entre el punto de cada cluster y cada punto del otro cluster.

Este método también se conoce como método de grupos de pares no ponderados con media aritmética.

* Pros y contras del método Average Linkage

**Ventajas del método Average Linkage**

* El método Average Linkage también funciona bien a la hora de separar clusters si hay ruido entre los clusters.

**Desventajas del método Average Linkage**

* El método Average Linkage está sesgado hacia los conglomerados globulares.

####  **Centroides**: basado en la media geométrica de todos los objetos

![](https://i0.wp.com/dataaspirant.com/wp-content/uploads/2020/12/13-Centroid-Linkage-Method.png?resize=768%2C492&ssl=1)

* En el enfoque Centroid Linkage, la distancia entre los dos conjuntos o clusters es la distancia entre dos vectores medios de los conjuntos (clusters).

En cada etapa, combinamos los dos conjuntos que tienen la distancia centroide más pequeña. En palabras sencillas, es la distancia entre los centroides de los dos conjuntos.

* Ventajas e inconvenientes del método Centroid Linkage

**Ventajas del método Centroid Linkage**

* El método Centroid Linkage también separa bien los conglomerados si hay ruido entre ellos.

**Desventajas del método Centroid Linkage**

* Al igual que los métodos Complete Linkage y Average Linkage, el método Centroid Linkage también está sesgado hacia los conglomerados globulares.

#### **Mediana**: basada en la mediana de todos los objetos

#### **Ward**: basado en la suma de cuadrados

![](https://i0.wp.com/dataaspirant.com/wp-content/uploads/2020/12/14-Wards-Linkage-Method.png?resize=768%2C570&ssl=1)

El método de vinculación de Ward es la similitud de dos conglomerados. Se basa en el aumento del error cuadrático cuando se fusionan dos conglomerados, y es similar a la media de grupo si la distancia entre puntos es la distancia al cuadrado.

* Pros y contras del método Ward's Linkage

**Ventajas del método Ward's Linkage**

* En muchos casos, se prefiere el método Ward's Linkage, ya que suele producir mejores jerarquías de conglomerados.
* El método de Ward es menos susceptible al ruido y a los valores atípicos.

**Desventajas del método Ward's Linkage**

* El método Ward's Linkage está sesgado hacia los clusters globulares.



### **Pasos del algorítmo para la agrupación jerárquica aglomerativa**

Sea X = {x1, x2, x3, ..., xn} el conjunto de puntos de datos.

1) Empezar con el clustering disjuntos que tiene nivel L(0) = 0 y número de secuencia m = 0.

2) Encuentre el par de conglomerados de menor distancia en la agrupación actual, digamos el par (r), (s), de acuerdo con **d[(r),(s)] = min d[(i),(j)]** donde el mínimo es sobre todos los pares de conglomerados en la agrupación actual.

3) Aumente el número de secuencia: m = m +1. **Combine los conglomerados (r) y (s) en un único conglomerado para formar el siguiente conglomerado m. Establezca el nivel de este conglomerado en L(m) = d[(r),(s)]**.

4) **Actualice la matriz de distancias, D, borrando las filas y columnas correspondientes a los clusters (r) y (s) y añadiendo una fila y una columna correspondientes al cluster recién formado**. La distancia entre el nuevo conglomerado, denominado (r,s) y el antiguo conglomerado(k) se define de la siguiente manera: **d[(k), (r,s)] = min (d[(k),(r)], d[(k),(s)]).**

5) Si todos los puntos de datos están en un clúster, deténgase; si no, repita desde el paso 2).

El **clustering Jerárquico Divisivo** es justo lo contrario del enfoque Jerárquico Aglomerativo.

* Ventajas

1) No hay información previa sobre el número de conglomerados necesarios.

2) Fácil de implementar y da el mejor resultado en algunos casos.

* Desventajas

1) El algoritmo nunca puede deshacer lo hecho anteriormente.

2) Se requiere una complejidad temporal de al menos O(n2 log n), donde "n" es el número de puntos de datos.

3) En función del tipo de matriz de distancias elegida para la fusión, los distintos algoritmos pueden sufrir uno o varios de los problemas siguientes

    i) Sensibilidad al ruido y a los valores atípicos.

    ii) rotura de conglomerados grandes

    iii) Dificultad para manejar conglomerados de diferentes tamaños y formas convexas.

4) No se minimiza directamente ninguna función objetivo

5) A veces es difícil identificar el número correcto de conglomerados mediante el dendograma.

### Generación de datos aleatorios
Estaremos generando un conjunto de datos usando la clase <b>make_blobs</b>. <br> <br>

<ul>

### Clustering Aglomerativo
Empezaremos por agrupar los puntos de datos aleatorios que acabamos de crear.

La clase <b> Agglomerative Clustering </b> requerirá dos entradas:
<ul>
    <li> <b>n_clusters</b>: el número de clústeres a formar, así como el número de centroides a generar. </li>
    <ul> <li> El valor será: 4 </li> </ul>
    <li> <b>enlace</b>: qué criterio de enlace utilizar. El criterio de vinculación determina qué distancia usar entre conjuntos de observación. El algoritmo fusionará los pares de clústeres que minimizan este criterio. </li>
    <ul>

    </ul>

Guarde el resultado en una variable llamada <b> agglom </b>

```python
agglom = AgglomerativeClustering(n_clusters = 4, linkage = 'single')
```

Ajuste el modelo con <b>X2</b> e <b>y2</b> a partir de los datos generados anteriormente.

```python
agglom.fit(X1,y1)
```

El siguiente código muestra la agrupación.

```python
# Create a figure of size 6 inches by 4 inches.
plt.figure(figsize=(6,4))

# These two lines of code are used to scale the data points down,
# Or else the data points will be scattered very far apart.

# Create a minimum and maximum range of X1.
x_min, x_max = np.min(X1, axis=0), np.max(X1, axis=0)

# Get the average distance for X1.
X1 = (X1 - x_min) / (x_max - x_min)

# This loop displays all of the datapoints.
for i in range(X1.shape[0]):
    # Replace the data points with their respective cluster value
    # (ex. 0) and is color coded with a colormap (plt.cm.spectral)
    plt.text(X1[i, 0], X1[i, 1], str(y1[i]),
             color=plt.cm.nipy_spectral(agglom.labels_[i] / 10.),
             fontdict={'weight': 'bold', 'size': 9})

# Remove the x ticks, y ticks, x and y axis
plt.xticks([])
plt.yticks([])
#plt.axis('off')



# Display the plot of the original data before clustering
plt.scatter(X1[:, 0], X1[:, 1], marker='.')
# Display the plot
plt.show()
```

### Dendrograma asociado para el agrupamiento jerárquico aglomerativo
Recordar que una **matriz de distancia** contiene la **distancia de cada punto a cualquier otro punto de un conjunto de datos**.

Utilizar la función `distance_matrix`, que requiere **dos entradas**. Usar la matriz de características, *X2* como ambas entradas y guardar la matriz de distancia en una variable llamada `dist_matrix`.

Recordar que los valores de distancia son simétricos, con una diagonal de 0's. Esta es una forma de asegurarse de que su matriz sea correcta. (imprimir `dist_matrix` para asegurarse de que sea correcto)

```python
dist_matrix = distance_matrix(X1,X1)
print(dist_matrix)
```

Utilizando la clase <b> linkage </b> de la jerarquía, pasar los parámetros:
<ul>
     <li>La matriz de distancia</li>
     <li> 'complete' para vinculación completa</li>
</ul>

Guardar el resultado en una variable llamada *Z*.

```python
Z = hierarchy.linkage(dist_matrix, 'single')
```

Un agrupamiento jerárquico generalmente se visualiza como un dendrograma, como se muestra en la siguiente celda. Cada combinación está representada por una línea horizontal. La coordenada y de la línea horizontal es la similitud de los dos grupos que se fusionaron, donde las ciudades se ven como grupos únicos.
Al pasar de la capa inferior al nodo superior, un dendrograma nos permite reconstruir el historial de fusiones que dieron como resultado el agrupamiento representado.

A continuación, guardaremos el dendrograma en una variable llamada <b>dendro</b>. Al hacer esto, también se mostrará el dendrograma.
Usando la clase <b>dendrogram</b> de la jerarquía, pasar el parámetro:

* *Z*

```python
dendro = hierarchy.dendrogram(Z)
```

# Clustering en un conjunto de datos de vehículos

Imaginemos que un fabricante de automóviles ha desarrollado prototipos de un nuevo vehículo. ### Limpieza de datos

Vamos a limpiar el conjunto de datos eliminando las filas que tienen valor nulo:

```python
print ("Shape of dataset before cleaning: ", pdf.size)
pdf[[ 'sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']] = pdf[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']].apply(pd.to_numeric, errors='coerce')
pdf = pdf.dropna()
pdf = pdf.reset_index(drop=True)
print ("Shape of dataset after cleaning: ", pdf.size)
pdf.head(5)
```

### Selección de características
Seleccionemos nuestro conjunto de características:

```python
featureset = pdf[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

### Normalización
Ahora podemos normalizar el conjunto de características. __MinMaxScaler__ transforma las características escalando cada característica a un rango dado. Por defecto es (0, 1). Es decir, este estimador escala y traduce cada característica individualmente de forma que esté entre cero y uno.

```python
from sklearn.preprocessing import MinMaxScaler
x = featureset.values #returns a numpy array
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx [0:5]
```

## Clustering usando **Scipy**
En esta parte utilizamos el paquete Scipy para agrupar el conjunto de datos:

[Scipy](https://scipy.org/)

En primer lugar, calculamos la matriz de distancias.

```python
import scipy

leng = feature_mtx.shape[0]
D = scipy.zeros([leng,leng])
for i in range(leng):
    for j in range(leng):
        D[i,j] = scipy.spatial.distance.euclidean(feature_mtx[i], feature_mtx[j])
```

```python
D
```

En el clustering aglomerativo, en cada iteración, el algoritmo debe actualizar la matriz de distancias para reflejar la distancia del cluster recién formado con el resto de clusters del bosque.
Los siguientes métodos son soportados en Scipy para calcular la distancia entre el cluster recién formado y cada uno:

* simple
* completo
* promedio
* centroide
* ward
  
Usamos __complete__ para nuestro caso, pero siéntete libre de cambiarlo para ver cómo cambian los resultados.

```python
import pylab
import scipy.cluster.hierarchy
Z = hierarchy.linkage(D, 'complete')
```

```python
dendro = hierarchy.dendrogram(Z)
```

Además, puede determinar directamente el número de clusters:

```python
from scipy.cluster.hierarchy import fcluster

k = 3
clusters = fcluster(Z, k, criterion='maxclust')
clusters
```

Ahora, graficamos el dendrograma:

```python
fig = pylab.figure(figsize=(18,50))
def llf(id):
    return '[%s %s %s]' % (pdf['manufact'][id], pdf['model'][id], int(float(pdf['type'][id])) )

dendro = hierarchy.dendrogram(Z,  leaf_label_func=llf, leaf_rotation=0, leaf_font_size=12, orientation='right')
```

## Clustering usando **scikit-learn**
Hagámoslo de nuevo, pero esta vez utilizando el paquete scikit-learn:

```python
dist_matrix = distance_matrix(feature_mtx,feature_mtx)
print(dist_matrix)
```

Ahora, podemos utilizar la función 'AgglomerativeClustering' de la biblioteca scikit-learn para agrupar el conjunto de datos. AgglomerativeClustering realiza una agrupación jerárquica utilizando un enfoque ascendente. El criterio de vinculación determina la métrica utilizada para la estrategia de fusión:

- Ward minimiza la suma de las diferencias al cuadrado dentro de todos los clusters. Es un enfoque de minimización de la varianza y, en este sentido, es similar a la función objetivo de k-means, pero abordada con un enfoque jerárquico aglomerativo.
- La vinculación máxima o completa minimiza la distancia máxima entre observaciones de pares de conglomerados.
- La vinculación media minimiza la media de las distancias entre todas las observaciones de pares de conglomerados.

```python
agglom = AgglomerativeClustering(n_clusters = 6, linkage = 'complete')
agglom.fit(feature_mtx)
agglom.labels_
```

Podemos añadir un nuevo campo a nuestro dataframe para mostrar el cluster de cada fila:

```python
pdf['cluster_'] = agglom.labels_
pdf.head()
```

```python
pdf["type"].value_counts("%")
```


# **Clustering no jerarquico**

## **K means**

Este metodo, como los de particiones en general, asume que el **numero
final de clusters (k) es conocido**. El procedimiento conlleva los siguientes
pasos:

1. Determinar los centroides iniciales o semillas de los k grupos.
2. Asignar cada observacion al centroide más cercano segun la distancia
(i.e. euclidea).
3. Con las observaciones clasificadas se recalculan los centroides de los k
grupos. Si las distancias entre los nuevos y viejos centroides es mayor que un criterio de convergencia establecido, se vuelve al paso 2.
4. El proceso termina cuando se cumple el criterio de convergencia.
Además, se fija un número de iteraciones que no debe superarse.

### Inercia

También conocida como suma de cuadrados dentro del cluster (WCSS por sus siglas en inglés, "within-cluster sum of squares"), mide la cantidad de dispersión (o varianza) de los puntos dentro de cada cluster. Específicamente, la inercia es la suma de las distancias al cuadrado de cada punto a su centroide más cercano.

Vamos a evaluar las particiones que realizamos mediante KMeans usando la **inercia**. La idea es que **al variar el número de clúster K en el modelo, el valor de la distancia media de los datos al centroide más cercano va a variar**. Queremos **graficar esa curva para elegir el número de particiones óptimos con el metodo del codo**.

Uno de los atributos de **kmeans en sklearn es inertia (pueden ver la doc)**:

image.png

La inercia tiene varias propiedades importantes:

* **Valor mínimo**: La inercia siempre es un valor no negativo.
* **Cero**: La inercia es cero cuando cada punto está exactamente en su centroide, lo cual es un caso ideal pero raramente alcanzable en la práctica.
* **Disminuye a medida que aumenta el número de clusters**: Si aumentas el número de clusters (K) al total de puntos (n) en tu conjunto de datos, cada punto se convertirá en su propio centroide y la inercia será cero.

image.png

Utilizando este atributo:

- Aplicar a el dataset 1 (X1) Kmeans con distintos valores de K: de 2 a 12.
- En cada iteración almacenar el resultado de inertia_ en una lista con el nombre **lista_inercias**

```python
K = np.arange(2,13)
```

```python
lista_inercias = []

for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X1)
    inercia = km.inertia_
    lista_inercias.append(inercia)
```

Ahora graficamos la inercia por cantidad de clusters:

```python
plt.figure(figsize = (10,7))
plt.plot(K,lista_inercias)
plt.scatter(K,lista_inercias,c='r')
plt.xlabel('Cantidad de Clusters K')
plt.ylabel('Inercia media')
plt.title('Método del codo para el Dataset 1')

plt.show()
```

¿Pueden ver el "codo"?

De esta forma para este problema **el número óptimo de clusteres sería 4.**


### Silhouette

Vamos a evaluar las particiones que realizamos mediante el valor de **silhouette**.

```python
from sklearn.metrics import silhouette_score
lista_sil = []

for k in range(2,13):
    km = KMeans(n_clusters=k)
    km = km.fit(X1)

    etiquetas = km.labels_

    valor_medio_sil = silhouette_score(X1, etiquetas)
    lista_sil.append(valor_medio_sil)

plt.figure(figsize = (10,7))
plt.plot(K, lista_sil, lw=3)
plt.scatter(K, lista_sil, c='r')
plt.xlabel('Cantidad de Clusters K')
plt.ylabel('Silhouette media')
plt.title('Silhouette media para el Dataset 2')

plt.show()
```

# **Maldición de la dimensionalidad**

**Los datos de alta dimensionalidad tienden a ser dispersos y los puntos están relativamente más distantes entre sí**. Los algoritmos que se basan en la distancia necesitan determinar la cercanía, pero más o menos dimensiones pueden sesgar los resultados.

Esto ocurre porque a medida que aumenta el número de dimensiones, la distancia media entre los puntos es mayor; a la inversa, a medida que disminuye el número de dimensiones, la distancia media es menor.

* Cuando nos enfrentamos a una tonelada de datos, podemos utilizar algoritmos de reducción de la dimensionalidad para que los datos "vayan al grano".

> **La maldición de la dimensionalidad describe la naturaleza explosiva del aumento de las dimensiones de los datos y el consiguiente incremento exponencial de los esfuerzos computacionales necesarios para su procesamiento y/o análisis**.

Este término fue introducido por primera vez por Richard E. Bellman para explicar el aumento del volumen del espacio euclidiano asociado a la adición de dimensiones extras en el ámbito de la programación dinámica. Hoy en día, este fenómeno se observa en campos como el aprendizaje automático, el análisis de datos o la minería de datos, por citar algunos. Un aumento de las dimensiones puede, en teoría, añadir más información a los datos, mejorando así su calidad, pero en la práctica aumenta el ruido y la redundancia durante su análisis.

# Introducción al análisis de componentes principales


## **Introducción**

**El Análisis de Componentes Principales (ACP)** se considera una de las técnicas más populares para la **extracción de características linealmente independientes y la reducción de la dimensionalidad**. A veces tenemos problemas de aprendizaje automático en los que las **características de entrada tienen dimensiones muy altas, lo que complica el aprendizaje, aumentando el procesamiento y reduciendo la precisión**.

Nuestra **primera tarea es reducir el espacio de características de entrada de alta dimensionalidad a un espacio de menor dimensionalidad que sea más eficaz en la tarea de aprendizaje**. El PCA tiene múltiples ventajas, como la compresión de datos, la mejora de la visualización, el aumento del rendimiento, la simplificación de los modelos, entre otras.

**Reducir la dimensionalidad del espacio de características de entrada conservando la mayor parte de la variabilidad de los datos es sólo uno de los posibles usos de PCA**. También es posible reconstruir los datos originales mediante la técnica de retroproyección.

## **Características**

* El análisis de componentes principales es un método rápido y flexible no supervisado para **reducir la dimensionalidad de los datos**.

* PCA es una técnica **solo aplicable a variables numéricas**, basandose en la variabilidad explicada por estas.

* Permite obtener un nuevo subconjunto de variables ordenadas en función de la cantidad de información que representan.

## **Objetivo de las componentes principales**

Permite pasar a un nuevo conjunto de variables (componentes principales)
**incorreladas** entre sí, ordenadas de acuerdo a la **informacion que incorporan**. A partir de estas es posible:

* **Reducir la dimension** de un fenómeno de interés (un menor número de variables), conservando la mayor cantidad información de las variables originales.
* **Encontrar nuevas estructuras de relaciones**
* **Utilizar las nuevas variables incorreladas en otras técnicas estadísticas**: regresión, cluster, etc. Permitiendo conservar la estructura de variabilidad.

![](https://bookdown.org/andreabellavia/mixtures/images/pca1.png)

### **Objetivo principal**

**Encontrar una proyección que maximice la varianza de los datos (centrados).** Es decir buscar la proyección de los datos en el sentido de mayor variabilidad de los mismos.

## **Importancia del escalado de características**

El **escalado de características** a través de la estandarización (o normalización Z-score) puede ser un paso importante del preprocesamiento para muchos algoritmos de aprendizaje automático. La estandarización consiste en reescalar las características para que tengan las propiedades de una distribución normal estándar con una media de cero y una desviación estándar de uno.

Mientras que muchos algoritmos (como SVM, K-vecinos más cercanos y regresión logística) **requieren que las características sean normalizadas**, intuitivamente podemos pensar en el **Análisis de Componentes Principales (PCA) como un otro ejemplo de cuando la normalización es necesaria**. **En el PCA nos interesan los componentes que maximizan la varianza**. Si un componente (por ejemplo, la altura humana) varía menos que otro (por ejemplo, el peso) debido a sus respectivas escalas (metros frente a kilos), **el ACP podría determinar que la dirección de la varianza máxima se corresponde más estrechamente con el eje del peso, si esas características no se escalan**. Dado que un cambio en la altura de un metro puede considerarse mucho más importante que el cambio en el peso de un kilogramo, esto es **claramente incorrecto**.

[Referencia](http://scikit-learn.org/stable/auto_examples/preprocessing/plot_scaling_importance.html)

PCA podría ser lo que necesita para desenredar ese lío de datos. Lo que hace es tomar los datos de expresión de 15 genes de cada ratón (todas las características) y **reducirlos a un único punto que representa el perfil de expresión de ese ratón**. Un punto para un ratón. Sesenta ratones, sesenta puntos. El resultado se verá así:

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_8.png)

Este gráfico facilita la comparación entre ratones. **Aquellos con perfiles de expresión similares se agruparán**. En este gráfico, hay 3 grupos de ratones. **A partir de aquí, se puede rastrear para averiguar qué genes hacen que los grupos sean diferentes entre sí (más sobre esto más adelante).**

Genial. ¿Cómo pasamos de esta tabla de 15×60 a este bonito gráfico?

No olvidemos que lo que intentamos hacer aquí, desde el momento en que llega la tabla de 15×60 y la miramos perplejos, es comparar los ratones. Cuando hay demasiada información sobre ellos, es difícil hacerlo. Empecemos con 2 genes para facilitar las cosas. Usando esos 2 genes como ejes, podemos trazar su expresión en 60 ratones en un gráfico 2D, así:

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_4c.png)

Aquí, cada punto lleva datos de lectura de 2 genes de un ratón, y juntos forman una "nube" plana. El componente principal 1 **(PC1)** es una línea que pasa por el centro de esa nube, captando el mayor porcentaje de variabilidad de la misma (explica al mayor cantidad de variabilidad de los datos **1**).

Es una línea que, si se proyectan sobre ella los puntos originales, ocurren dos cosas:

* La **distancia total entre los puntos proyectados es máxima**. Esto significa que se pueden **distinguir unos de otros con la mayor claridad posible**. Queremos comparar cosas, ¿recuerdas? Si una línea mezcla los puntos de datos entre sí, no servirá de nada.
* **La distancia total entre los puntos originales y sus correspondientes puntos proyectados es mínima**. Esto significa que tenemos una **representación lo más cercana posible a los datos originales**.

En otras palabras, **la mejor línea -nuestra PC1- debe transmitir la máxima variación entre los puntos de datos y contener el mínimo error**.

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_6.png)

**2. El ACP aborda la maldición de la dimensionalidad capturando la esencia de los datos en un reducido numero de componentes principales.**

Pero tenemos 15 genes, no sólo 2. Cuantos más genes tengamos, más ejes (dimensiones) habrá cuando representemos su expresión. Cuando lleguemos al cuarto gen, probablemente ya no querremos representar esto, y mucho menos 15 genes. **15 genes equivalen a 15 ejes. La nube de puntos ya no es plana, ahora tiene 15 dimensiones. No hay manera de mirarla y encontrarle sentido.**

Aquí es cuando entra la magia del PCA. **Para crear PC1, se ancla una línea en el centro de la nube de puntos 15-D y se gira en 15 direcciones, actuando al mismo tiempo como un "espejo" sobre el que se proyectan los 60 puntos originales. Esta rotación continúa hasta que la distancia total entre los puntos proyectados es máxima. La línea giratoria describe ahora la mayor variación entre los 60 ratones, y se ajusta para ser PC1.**

![](https://cdn-images-1.medium.com/max/800/0*XqrkXQoEQEIRe0Nm)

PC2 es la segunda línea que se encuentra con PC1, perpendicularmente (**no correlacionadas**), en el centro de la nube, y **describe la segunda dirección de mayor variación en los datos**. (Los ordenadores pueden manejar toda esta rotación y establecer PC en un santiamén, ¡qué suerte!).

**Si el ACP es adecuado para sus datos, sólo los 2 ó 3 primeros componentes principales deberían contener la mayor parte de la información de los datos** (más sobre esto más adelante). Esto es bueno en varios sentidos:

* Los componentes principales ayudan a reducir el número de dimensiones a 2 ó 3, lo que permite patrones.
* Sin embargo, no hay que desechar ningún gen al hacerlo. Los componentes principales **tienen en cuenta todas las dimensiones y puntos de datos**.
* Como PC1 y PC2 son perpendiculares entre sí, podemos rotarlos y hacerlos rectos. Estos son los ejes de nuestro bonito gráfico PCA.

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_6b.png)

**3. Las dimensiones varían en el peso que tienen en cada componente principal.**

Ahora que tenemos los ejes, ¿qué hay de los puntos? ¿Cómo unimos 15 genes para crear un único punto para cada ratón?

Los puntos originales (expresión génica del ratón) se fijan en su lugar por sus valores en todos los ejes (15 genes). Estos puntos se proyectan sobre las líneas PC, que intentan acercarse lo más posible a todos los puntos; así es como consiguen el mejor ajuste. Para las componentes, esto es como intentar quedarse quietas mientras se tira de ellas en 15 direcciones a la vez. Algunas fuerzas de tracción son más fuertes que otras; en otras palabras, **las dimensiones tienen distinto peso a la hora de definir cada PC.**

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_loadingplot.png)

En nuestro ejemplo, los 15 genes participan en la determinación de los ratones en los PC, pero su peso es diferente. Una vez determinados sus pesos, averiguar dónde colocan al ratón nº 1 en el PC1 es una simple operación matemática:

> (dato gen1 * peso gen1 en PC1)+
(dato gen2 * peso gen2 en PC1)+
...+
(dato gen15 * peso gen15 en PC1)
= valor PC1 del ratón nº 1

El ratón nº 1 también tiene un punto en PC2. Para calcularlo, utilizamos la misma fórmula con pesos de genes en PC2 en lugar de PC1.

Con un valor de PC1 y un valor de PC2, el Ratón #1 ahora puede ser graficado como un punto en el gráfico PCA.

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_1.png)

Repitiendo esta operación con todos los ratones, cada uno de ellos se convertirá en un punto en el gráfico PCA.

![](https://blog.bioturing.com/wp-content/uploads/2018/11/Blog_pca_8.png)

¡Qué bonito! ¿Cómo se lee esto?

**4. Cómo leer un gráfico PCA**

Los **ratones que tienen perfiles de expresión similares están ahora agrupados**. Sólo con echar un vistazo a este gráfico, podemos ver que hay **3 grupos de ratones**.
**Si 2 clusters de ratones son diferentes basados en PC1, como los clusters azul y naranja en este gráfico, tales diferencias son probablemente debidas a los genes que tienen una fuerte influencia en PC1.**
Si 2 grupos son diferentes en función de PC2, como los grupos rojo y azul, entonces es probable que los genes que influyen fuertemente en PC2 sean los responsables.
**Tenga en cuenta que las PC se clasifican según el grado en que describen los datos**. PC1 revela la mayor variación, mientras que PC2 revela la segunda mayor variación. Por lo tanto, las diferencias entre clusters a lo largo del eje PC1 son realmente mayores que las distancias similares a lo largo del eje PC2.

¿Tiene sentido este gráfico? Compruebe la proporción de varianza, o el diagrama de diagnóstico. **PCA merece la pena si las 2 ó 3 primeras PC cubren la mayor parte de la variación de sus datos**.

![](https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/national/Principal%2520Component%2520Analysis%2520Principal%2520Components.png)

Gráficos de porcentaje de varianza.

En resumen, el análisis de componentes principales (ACP) permite extraer patrones sólidos de conjuntos de datos grandes y complejos. **La esencia de los datos se capta en unas pocas componentes principales, que transmiten el mayor porcentaje de la variación del conjunto de datos**. El ACP reduce el número de dimensiones sin seleccionarlas ni descartarlas. En su lugar, **construye componentes principales que se centran en la variación y tienen en cuenta las distintas influencias de las dimensiones**. Dichas influencias pueden rastrearse a partir del gráfico PCA para averiguar qué produce las diferencias entre los conglomerados.

## Detalles matemáticos

**Geométricamente**: los ejes originales son transformados efectuando primero una **traslación** del origen al centroide, y luego una **rotación** que determina los nuevos ejes en el sentido de mayor variabilidad.

La obtención de las componentes se basa en encontrar los **autovectores** y **autovalores** a la matriz de covarianza de las features.

* **autovectores** -> componente principal.

* **autovalores** -> % de variabilidad explicado por cada componente.

## **Ejemplo 2**: Set de datos de vinos

El conjunto de datos sobre vino consta de 13 parámetros diferentes, como el contenido de alcohol y cenizas, medidos en 178 muestras de vino.

Estos vinos se cultivaron en la misma región de Italia, pero procedían de tres variedades diferentes, por lo que hay tres clases distintas de vino.

El objetivo es encontrar un modelo que pueda predecir la clase de vino a partir de los 13 parámetros medidos y averiguar las principales diferencias entre las tres clases.

A continuación procederemos a comparar un modelo de aprendizaje automatico con las componentes principales creadas a partir de los datos originales y con las componentes principales a partir de los datos estandarizados.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.datasets import load_wine
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Load the sample data
features, target = load_wine(return_X_y=True)
```

```python
# Make a train/test split using 30% test size
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30)

# Fit to data and predict using pipelined predict and PCA.
unscaled_clf = make_pipeline(PCA(n_components=2), LogisticRegression())
unscaled_clf.fit(X_train, y_train)
pred_test = unscaled_clf.predict(X_test)

# Fit to data and predict using pipelined scaling, predict and PCA.
std_clf = make_pipeline(StandardScaler(), PCA(n_components=2), LogisticRegression())
std_clf.fit(X_train, y_train)
pred_test_std = std_clf.predict(X_test)

# Show prediction accuracies in scaled and unscaled data.
norm_pred_accuracy = metrics.accuracy_score(y_test, pred_test)
std_pred_accuracy = metrics.accuracy_score(y_test, pred_test_std)
print(f'Prediction accuracy for the normal test dataset with PCA: {norm_pred_accuracy}')
print(f'Prediction accuracy for the standardized test dataset with PCA: {std_pred_accuracy}')
```

```python
pca = PCA(n_components = 13)
model = pca.fit(features)
model.components_
```

Podemos notar que el modelo de clasificación aumenta su desempeño cuando se ajusta a partir de las componentes principales con los datos escalados.

## **Ejemplo 5**: Mayor profundidad al análisis de componentes principales

El PCA es fundamentalmente un algoritmo de reducción de la dimensionalidad, pero también puede ser útil como herramienta de visualización, filtrado de ruido, extracción de características e ingeniería, y mucho más.
Tras una breve discusión conceptual del algoritmo PCA, exploraremos un par de ejemplos de estas aplicaciones adicionales.

El análisis de componentes principales es un método rápido y flexible no supervisado para la reducción de dimensionalidad en los datos, que vimos brevemente en [Introducing Scikit-Learn](05.02-Introducing-Scikit-Learn.ipynb).
Su comportamiento es más fácil de visualizar observando un conjunto de datos bidimensional.
Consideremos estos 200 puntos (véase la siguiente figura):

```python
import numpy as np
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

A simple vista, está claro que existe una relación casi lineal entre las variables *x* e *y*.
Esto recuerda a la Regresión lineal, pero la configuración del problema aquí es ligeramente diferente: en lugar de intentar *predecir* los valores *y* a partir de los valores *x*, **el problema de aprendizaje no supervisado intenta aprender sobre la *relación* entre los valores *x* e *y*.**

En el análisis de componentes principales, esta relación se cuantifica encontrando una lista de los **ejes principales** en los datos, y usando esos ejes para describir el conjunto de datos.

Usando el estimador `PCA` de Scikit-Learn, podemos calcular esto como sigue:

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

El ajuste aprende algunas cantidades de los datos, sobre todo los componentes y la explicación de la varianza para cada uno:

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

Para ver qué significan estas cifras, visualicémoslas como vectores sobre los datos de entrada, utilizando los componentes para definir la dirección del vector y la varianza explicada para definir la longitud al cuadrado del vector (véase la figura siguiente):

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->', linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

Estos vectores representan los ejes principales de los datos, y la longitud de cada vector es una medida de "importancia" de ese eje a la hora de describir la variabilidad de los datos, es una medida de la varianza captada por el eje cuando los datos cuando se proyectan sobre el mismo.

La proyección de cada punto de datos sobre los ejes principales son los componentes principales de los datos.

Esta transformación de ejes de datos a ejes principales es una **transformación afín**, lo que significa que se compone de una traslación, una rotación y un escalado uniforme.

Aunque este algoritmo para hallar componentes principales puede parecer solo una curiosidad matemática, resulta que tiene aplicaciones de muy largo alcance en el mundo del aprendizaje automático y la exploración de datos.

### El ACP como reducción de la dimensionalidad

El uso de PCA para la reducción de la dimensionalidad implica la eliminación de uno o más de los componentes principales más pequeños, lo que resulta en una proyección de menor dimensión de los datos que preserva la varianza máxima de los datos.

He aquí un ejemplo de utilización del PCA como transformada de reducción de la dimensionalidad:

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```

Los datos transformados se han reducido a una sola dimensión.
Para comprender el efecto de esta reducción de la dimensionalidad, podemos realizar la transformada inversa de estos datos reducidos y representarlos gráficamente junto con los datos originales (véase la figura siguiente):

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```

Los puntos claros son los datos originales, mientras que los oscuros son la versión proyectada.
Esto aclara lo que significa una reducción de dimensionalidad PCA: se elimina la información a lo largo del eje o ejes principales menos importantes, dejando sólo el componente o componentes de los datos con la mayor varianza.
La fracción de varianza que se elimina (proporcional a la dispersión de los puntos en torno a la línea formada en la figura anterior) es aproximadamente una medida de cuánta "información" se descarta en esta reducción de la dimensionalidad.

Este conjunto de datos de dimensión reducida es, en cierto sentido, "suficientemente bueno" para codificar las relaciones más importantes entre los puntos: a pesar de reducir el número de características de los datos en un 50%, las relaciones generales entre los puntos de datos se conservan en su mayor parte.

### ¿Qué significan los componentes?

Podemos ir un poco más lejos y empezar a preguntarnos qué *significan* las dimensiones reducidas.
Este significado puede entenderse en términos de combinaciones de vectores base.
Por ejemplo, cada imagen del conjunto de entrenamiento está definida por una colección de 64 valores de píxeles, que llamaremos el vector $x$:

$$
x = [x_1, x_2, x_3 \cdots x_{64}]
$$

Una forma de pensar en esto es en términos de una base de píxeles.
Es decir, para construir la imagen, multiplicamos cada elemento del vector por el píxel que describe y, a continuación, sumamos los resultados para construir la imagen:

$$
{\rm image}(x) = x_1 \cdot{\rm (pixel~1)} + x_2 \cdot{\rm (pixel~2)} + x_3 \cdot{\rm (pixel~3)} \cdots x_{64} \cdot{\rm (pixel~64)}
$$

Una forma de reducir la dimensionalidad de estos datos es eliminar todos estos vectores base excepto algunos.
Por ejemplo, si utilizamos sólo los ocho primeros píxeles, obtenemos una proyección de ocho dimensiones de los datos. Sin embargo, no refleja bien toda la imagen: ¡hemos eliminado casi el 90% de los píxeles!

image

La fila superior de paneles muestra los píxeles individuales, y la fila inferior muestra la contribución acumulativa de estos píxeles a la construcción de la imagen.
Utilizando sólo ocho de los componentes de la base de píxeles, sólo podemos construir una pequeña parte de la imagen de 64 píxeles.
Si continuáramos esta secuencia y utilizáramos los 64 píxeles, recuperaríamos la imagen original.

Pero la representación por píxeles no es la única elección de base. También podemos utilizar otras funciones de base, cada una de las cuales contiene alguna contribución predefinida de cada píxel, y escribir algo como:

$$
imagen(x) = {\rm media} + x_1 \cdot{\rm (base~1)} + x_2 \cdot{\rm (base~2)} + x_3 \cdot{\rm (base~3)} \cdots
$$

El ACP puede concebirse como un proceso de elección de funciones de base óptimas, de forma que la suma de solo las primeras de ellas baste para reconstruir adecuadamente el grueso de los elementos del conjunto de datos.
Las componentes principales, que actúan como la representación de baja dimensión de nuestros datos, son simplemente los coeficientes que multiplican cada uno de los elementos de esta serie.
La siguiente figura muestra una representación similar de la reconstrucción del mismo dígito utilizando la media más las ocho primeras funciones de base PCA:

imagen.png

A diferencia de la base de píxeles, la base PCA nos permite recuperar las características más destacadas de la imagen de entrada con sólo una media, ¡más ocho componentes!
La cantidad de cada píxel en cada componente es el corolario de la orientación del vector en nuestro ejemplo bidimensional.
Este es el sentido en el que PCA proporciona una representación de baja dimensión de los datos: descubre un conjunto de funciones base que son más eficientes que la base nativa de píxeles de los datos de entrada.

### Elección del número de componentes

Una parte vital del uso del ACP en la práctica es la capacidad de estimar cuántos componentes son necesarios para describir los datos.
Esto puede determinarse observando la *razón de varianza explicada* acumulada en función del número de componentes (véase la siguiente figura):

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

Esta curva cuantifica qué parte de la varianza total de 64 dimensiones está contenida en los primeros $n$ componentes.
Por ejemplo, vemos que con los datos de los dígitos los 10 primeros componentes contienen aproximadamente el 75% de la varianza, mientras que se necesitan unos 50 componentes para describir cerca del 100% de la varianza.

Esto nos indica que nuestra proyección bidimensional pierde mucha información (medida por la varianza explicada) y que necesitaríamos unos 20 componentes para conservar el 90% de la varianza.  Observar este gráfico para un conjunto de datos de alta dimensión puede ayudar a comprender el nivel de redundancia presente en sus características.

## **Ejemplo 7**: PCA como filtro de ruido

El ACP también puede utilizarse como método de filtrado de datos ruidosos.
La idea es la siguiente: **cualquier componente con una varianza mucho mayor que el efecto del ruido debería verse relativamente poco afectado por el ruido.
Por lo tanto, si reconstruye los datos utilizando sólo el subconjunto más grande de componentes principales, debería conservar preferentemente la señal y desechar el ruido.**

Veamos cómo funciona esto con los datos de los dígitos.

Primero trazaremos varias de las muestras de entrada sin ruido (la siguiente figura):

```python
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(digits.data)
```

Ahora vamos a **añadir un poco de ruido aleatorio** para crear un conjunto de datos ruidoso y volver a trazarlo (figura siguiente):

```python
rng = np.random.default_rng(42)
rng.normal(10, 2)
```

```python
rng = np.random.default_rng(42)
noisy = rng.normal(digits.data, 4)
plot_digits(noisy)
```

La visualización deja clara la presencia de este ruido aleatorio.
Entrenemos un modelo PCA sobre los datos ruidosos, solicitando que la proyección preserve el 50% de la varianza:

```python
pca = PCA(0.50).fit(noisy)
pca.n_components_
```

En este caso, el 50% de la varianza corresponde a 12 componentes principales de las 64 características originales.
Ahora calculamos estos componentes y, a continuación, utilizamos la inversa de la transformada para reconstruir los dígitos filtrados; la siguiente figura muestra el resultado:

```python
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

**Esta propiedad de preservar la señal y filtrar el ruido convierte al PCA en una rutina de selección de características muy útil**: por ejemplo, en lugar de entrenar un clasificador en datos de muy alta dimensión, se puede entrenar el clasificador en la representación de componentes principales de menor dimensión, que servirá automáticamente para filtrar el ruido aleatorio de las entradas.

## **Ejemplo 8**:  Eigenfaces

Conjunto de datos Labeled Faces in the Wild (LFW) disponible a través de Scikit-Learn:

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

Echemos un vistazo a los ejes principales que abarcan este conjunto de datos.
Dado que se trata de un gran conjunto de datos, **utilizaremos el eigensolver** "aleatorio" en el estimador "PCA": **utiliza un método aleatorio para aproximar los primeros $n$ componentes principales más rápidamente que el enfoque estándar, a expensas de cierta precisión**. Esta compensación puede ser útil para datos de alta dimensionalidad (en este caso, una dimensionalidad de casi 3.000).

Analizaremos los primeros 150 componentes:

```python
pca = PCA(150, svd_solver='randomized', random_state=42)
pca.fit(faces.data)
```

En este caso, puede ser interesante visualizar las imágenes asociadas a los primeros componentes principales (estos componentes se conocen técnicamente como *vectores propios*,
por lo que este tipo de imágenes suelen denominarse *eigenfaces*; como puede verse en la siguiente figura, son tan espeluznantes como suenan):

```python
fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
```

Los resultados son muy interesantes y nos dan una idea de cómo varían las imágenes: por ejemplo, los primeros eigenfaces (desde arriba a la izquierda) parecen estar asociados con el ángulo de iluminación de la cara, y los vectores principales posteriores parecen estar recogiendo ciertos rasgos, como ojos, narices y labios.
Echemos un vistazo a la varianza acumulada de estos componentes para ver qué parte de la información de los datos conserva la proyección (véase la figura siguiente):

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

Los 150 componentes que hemos elegido explican algo **más del 90% de la varianza.**

Esto nos llevaría a pensar que, utilizando estos 150 componentes, recuperaríamos la mayor parte de las características esenciales de los datos.
Para concretarlo, podemos comparar las imágenes de entrada con las imágenes reconstruidas a partir de estos 150 componentes (véase la figura siguiente):

```python
# Compute the components and projected faces
pca = pca.fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)
```

```python
# Plot the results
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')

ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction');
```

La fila superior muestra las imágenes de entrada, mientras que la inferior muestra la reconstrucción de las imágenes a partir de **sólo 150 de las ~3.000 características iniciales.**

Esta visualización deja claro por qué la selección de características PCA utilizada en [In-Depth: Support Vector Machines](05.07-Support-Vector-Machines.ipynb) tuvo tanto éxito: **aunque reduce la dimensionalidad de los datos en casi un factor de 20, las imágenes proyectadas contienen suficiente información como para que podamos, a ojo, reconocer a los individuos en cada imagen**. Esto significa que **nuestro algoritmo de clasificación sólo necesita entrenarse con datos de 150 dimensiones en lugar de con datos de 3.000 dimensiones, lo que, dependiendo del algoritmo concreto que elijamos, puede dar lugar a una clasificación mucho más eficiente.**

# Selección de características (feature selection)

La selección de características es un proceso en el que se seleccionan automáticamente las características de los datos que más contribuyen a la variable de predicción o al resultado en el que se está interesado.

La presencia de características irrelevantes en los datos puede reducir la precisión de muchos modelos, especialmente los algoritmos lineales como la regresión lineal y logística.

La selección de características antes de modelar los datos tiene tres ventajas:

* **Reduce el sobreajuste**: Menos datos redundantes significa menos oportunidades de tomar decisiones basadas en el ruido.
* **Mejora la precisión**: Menos datos engañosos significa que mejora la precisión del modelado.
* **Reduce el tiempo de entrenamiento**: menos datos significa que los algoritmos se entrenan más rápido.

Más información sobre la selección de características con scikit-learn en su [documentación](https://scikit-learn.org/stable/modules/feature_selection.html).

## Selección univariada

La selección de características univariadas consiste en seleccionar las mejores características basándose en pruebas estadísticas univariada. Puede considerarse como un paso previo al procesamiento de un estimador.

[Documentación de sci-kit learn](https://scikit-learn.org/stable/modules/feature_selection.html#univariate-feature-selection)

```python
# Make a train/test split using 30% test size
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30)

# Fit to data and predict using pipelined scaling, predict and PCA.
std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X_train, y_train)

X_scaled.shape
```

```python
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif  # ANOVA F-value
 # otras opciones posibles: chi2, mutual_info_classif
 # Para regresión: r_regression, f_regression, mutual_info_regression

# feature extraction
selector = SelectKBest(score_func=f_classif, k=4)
fit = selector.fit(X_scaled, y_train)

# imprimimos los scores
set_printoptions(precision=3)
print("Puntajes:", fit.scores_)

X_scaled_reduced = fit.transform(X_scaled)
# imprimimos X con los 4 features seleccionados
print("X reducido:", X_scaled_reduced[0:5,:])
```

## Recursive Feature Elimination

La Eliminación Recursiva de Características (o RFE) consiste en eliminar recursivamente atributos y construir un modelo sobre los atributos que quedan. Utiliza la precisión del modelo para identificar qué atributos (y combinación de atributos) contribuyen más para predecir el atributo objetivo.

[Documentación de sci-kit learn](http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html#sklearn.feature_selection.RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# feature extraction
model = LogisticRegression(solver='lbfgs')
rfe = RFE(model, n_features_to_select=3)
fit = rfe.fit(X_scaled, y_train)

print("Número de features:", fit.n_features_)
print("Features seleccionadas:", fit.support_)
print("Ranking:", fit.ranking_)
```
